```python
from src.preprocessing.denoise import filter_noise

def test_filter_noise():
    # Event (polarity, row, col, timestamp)
    test_events = [
        0, 5, 0, 100,   
        1, 5, 1, 101,    
        0, 5, 6, 102,   
        1, 5, 9, 103,  
        1, 5, 12, 104
    ]

    events = [test_events]

    filtered = filter_noise(events, img_width=20, img_height=20, box_size=7, threshold=3)

    print("Original events:", len(test_events) // 4, "events")
    print("Filtered events:", len(filtered[0]) // 4, "events")
    print("\nFiltered events details:")
    for i in range(0, len(filtered[0]), 4):
        print(f"  Event at ({filtered[0][i+1]}, {filtered[0][i+2]})")

    assert len(filtered[0]) == 4, f"Expected 4 params, got {len(filtered[0])}"
    print("\n✓ Test passed! Isolated event was filtered out, cluster was kept.")

test_filter_noise()
```

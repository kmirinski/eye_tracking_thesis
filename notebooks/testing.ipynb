```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from matplotlib.patches import Polygon
import numpy as np

W, H = 346, 260
size = 80

fig, axes = plt.subplots(1, 2, figsize=(10, 4))

for ax, corner, verts, title in [
    (axes[0], 'upper_right',
    [(W - size, 0), (W, 0), (W, size)],
    'upper_right (left eye)'),
    (axes[1], 'upper_left',
    [(0, 0), (size, 0), (0, size)],
    'upper_left (right eye)'),
]:
    ax.set_xlim(0, W)
    ax.set_ylim(H, 0)  # flip y so (0,0) is top-left
    ax.set_aspect('equal')
    ax.add_patch(Polygon(verts, closed=True, color='red', alpha=0.5, label='excluded'))
    ax.set_title(title)
    ax.set_xlabel('cx (px)')
    ax.set_ylabel('cy (px)')
    ax.legend()

plt.tight_layout()
plt.show()
```
